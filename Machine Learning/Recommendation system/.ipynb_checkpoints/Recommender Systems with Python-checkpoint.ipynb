```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id','item_id','rating','timestamp']
df = pd.read_csv('u.data',sep = '\t',names = column_names)
```

```python
df
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
movie_titles
```

```python
df = pd.merge(df,movie_titles,on = 'item_id')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending = False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending = False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize = (10,4))
ratings['num of ratings'].hist(bins = 70)
```

```python
sns.histplot(ratings,x = 'num of ratings')
plt.ylabel('')
```

```python
plt.figure(figsize = (10,4))
ratings['rating'].hist(bins = 70)
```

```python
sns.jointplot(data = ratings,x = 'rating',y = 'num of ratings')
```

```python
moviesmat = df.pivot_table(index='user_id',columns = 'title',values = 'rating')
moviesmat.head()
```

```python
ratings.sort_values('num of ratings',ascending = False).head(10)
```

```python
starwars_user_ratings = moviesmat['Star Wars (1977)']
liarliar_user_ratings = moviesmat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviesmat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviesmat.corrwith(liarliar_user_ratings)
```

```python
similar_to_starwars
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns = ['Correlation'])
corr_starwars.dropna(inplace = True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending = False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending = False).head(10)
```
