### Data Cleaning

```python
plt.figure(figsize=(12, 7))
sb.boxplot(x='Pclass',y='Age',data=titanic,palette='winter')
```

```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 17
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age

```

```python
titanic['Age']=titanic[['Age','Pclass']].apply(impute_age,axis=1)
sb.heatmap(titanic.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```
