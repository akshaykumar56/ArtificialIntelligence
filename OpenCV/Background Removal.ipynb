```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
mg=cv2.imread('butterfly.jpg')
new_img=cv2.cvtColor(mg, cv2.COLOR_RGB2BGR)
hsv=cv2.cvtColor(new_img, cv2.COLOR_BGR2HSV)
lower_green=np.array([25, 52, 72])
upper_green=np.array([102, 255, 255])


mask2=cv2.inRange(hsv, lower_green, upper_green)


res2=cv2.bitwise_and(new_img, new_img, mask=mask2)
plt.imshow(new_img)


```

```python
plt.imshow(res2)
```

```python
cv2.imwrite("totally_new.jpg",res2)
```

```python
mg=cv2.imread('totally_new.jpg')
new_img=cv2.cvtColor(mg, cv2.COLOR_RGB2BGR)
hsv=cv2.cvtColor(new_img, cv2.COLOR_BGR2HSV)
lower_black=np.array([0, 0, 0])
upper_black=np.array([180, 255, 30])


mask3=cv2.inRange(hsv, lower_black, upper_black)


res3=cv2.bitwise_and(new_img, new_img, mask=mask3)
plt.imshow(new_img)
```
