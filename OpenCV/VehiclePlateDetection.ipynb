```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
#cascade classifier
cpath=os.path.dirname(cv2.__file__)+"/data/haarcascade_russian_plate_number.xml"
face=cv2.CascadeClassifier(cpath)
img=cv2.imread("car2.jfif")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces=face.detectMultiScale(gray)
print(type(faces))
```

```python
if len(faces)==0:
    print("no number plates found!")
else:
    print(faces)
    print(faces.shape)
    print("number plate detected:"+str(faces.shape[0]))
    for (x,y,w,h) in faces:
        d=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),1)
cv2.putText(img,"no. plate detected"+str(faces.shape[0]),(0,img.shape[0] -10),cv2.FONT_HERSHEY_TRIPLEX,1,(255,0,0))
cv2.waitKey(0)
cv2.destroyAllWindows()
# cv2.imwrite('img2.jpg',d)
```
