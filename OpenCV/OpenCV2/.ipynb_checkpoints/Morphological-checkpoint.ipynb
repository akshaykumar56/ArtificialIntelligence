```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
gray_img=cv2.imread('aksh.jpg',cv2.IMREAD_GRAYSCALE)
kernel = np.ones((6,6),np.uint8)
```

```python
er=cv2.erode(gray_img,kernel,iterations=1)
```

```python
plt.subplot(121),plt.imshow(gray_img),plt.title("original")
plt.subplot(122),plt.imshow(er),plt.title("Eroded Image")
```

```python
white=cv2.imread('D.jpg')
kernel = np.ones((4,4),np.uint8)
d1=cv2.dilate(white,kernel,iterations=1)
plt.subplot(121),plt.imshow(white),plt.title('original')
plt.subplot(122),plt.imshow(d1),plt.title('Dilation Image')
```

```python
#1. Erosion
# Erosion primarily involves eroding the outer surface (the foreground) of the image. As binary images only contain two pixels 
# 0 and 255, it primarily involves eroding the foreground of the image and it is suggested to have the foreground as white. 
# The thickness of erosion depends on the size and shape of the defined kernel. We can make use of NumPy’s ones() function to 
# define a kernel. There are a lot of other functions like NumPy zeros, customized kernels, and others that can be used to 
# define kernels based on the problem in hand.

img = cv2.imread(r"E.jpg",0)
# binarize the image
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
 
# define the kernel
kernel = np.ones((10, 10), np.uint8)
 
# invert the image
invert = cv2.bitwise_not(binr)
 
# erode the image
erosion = cv2.erode(invert, kernel,iterations=1)
 
# print the output
plt.imshow(erosion, cmap='gray')

```

```python
plt.imshow(img, cmap='gray')
```

```python
plt.imshow(invert, cmap='gray')
```

### Dilation

```python
# Dilation involves dilating the outer surface (the foreground) of the image. 
# As binary images only contain two pixels 0 and 255, it primarily involves expanding the foreground of the image 
# and it is suggested to have the foreground as white. The thickness of erosion depends on the size and 
# shape of the defined kernel. We can make use of NumPy’s ones() function to define a kernel.
# There are a lot of other functions like NumPy zeros, customized kernels, and others that can be used to define kernels 
# based on the problem at hand. It is exactly opposite to the erosion operation

img = cv2.imread(r"E.jpg", 0)
 
# binarize the image
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
 
# define the kernel
kernel = np.ones((3, 3), np.uint8)
 
# invert the image
invert = cv2.bitwise_not(binr)
 
# dilate the image
dilation = cv2.dilate(invert, kernel, iterations=3)
 
# print the output
plt.imshow(dilation, cmap='gray')
```

```python
# Morphological gradient is slightly different than the other operations, because, the morphological gradient first 
# applies erosion and dilation individually on the image and then computes the difference between the eroded and dilated image. 
# The output will be an outline of the given image
img = cv2.imread(r"E.jpg", 0)
 
# binarize the image
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
# define the kernel
kernel = np.ones((3, 3), np.uint8)
# invert the image
invert = cv2.bitwise_not(binr)
# use morph gradient
morph_gradient = cv2.morphologyEx(invert,cv2.MORPH_GRADIENT,kernel)
 
# print the output
plt.imshow(morph_gradient, cmap='gray')
```
