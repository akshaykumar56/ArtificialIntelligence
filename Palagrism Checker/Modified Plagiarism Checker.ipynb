```python
# Plagiarism Checker

import tkinter as tk
from tkinter import *
from tkinter import filedialog

import pandas as pd

import math

from tkinter.filedialog import askopenfile
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
global File1
global File2
window = tk.Tk()  
window.state('zoomed')
window.title('Plagiarism Checker')
font1=('times', 18, 'bold')

label1 = tk.Label(window,text='Plagiarism Checker',width=30,font=font1)  
label1.place(x=650,y=70,anchor='n')

label2 = tk.Label(window,text='Upload File',width=30,font=font1)  
label2.place(x=650,y=110,anchor='n')

label3 = tk.Label(window,text='Check for Plagiarism with Our DataBase',width=30,font=font1)  
label3.place(x=650,y=140,anchor='n')

# Function to Draw Progress Bar

def getProgressAsStringProgressBar(progress):
    progressBar=str()
    for i in range(10):
        if(i<int(progress * 10)+1):    
            progressBar += '█'
        else:
            progressBar += '░'
    progressBar +='\t' + str(math.trunc(progress * 100)) + '%'
    
    return progressBar

# Uses as a DataBase File--> Change the Below Path to the path of file where you store your DataBase File

fp = open(r'E:\plagiarism-checker-master\DataBase File.txt', 'r',encoding='utf-8')
dataBase=fp.read()
# print(dataBase)

# File That You Want To Check For Plagiarism

def upload_file2():
    file = filedialog.askopenfilename(filetypes = [('All Files', '*.*'), 
             ('PDF Files', '*.pdf'),
             ('Text Document', '*.txt'),
              ('CSV files',"*.csv")])
    fob=open(file,'r',encoding='utf-8')
    File2=fob.read()

# Function To Convert Text File Into Vectors

def vectorize(text):
    return TfidfVectorizer().fit_transform(text).toarray()

# Function To Convert Text File Into Vectors

def similarity(doc1,doc2):
    return cosine_similarity([doc1,doc2])


df1 = pd.DataFrame({'First_File': [dataBase], 'Second_File': [File2]})
print(df1)
tfidf_vectorizer = TfidfVectorizer()

doc_vec = tfidf_vectorizer.fit_transform(df1.iloc[0])
doc_vec


df2 = pd.DataFrame(doc_vec.toarray().transpose(),index=tfidf_vectorizer.get_feature_names())
df2.columns = df1.columns


z=['First File','Second File']
vectors=vectorize(df1.iloc[0])
# print(vectors)

s_vectors=list(zip(z,vectors))


plagiarism_results=set()

# check_plagiarism --> Main Function

def check_plagiarism():
    for student_a,text_vector_a in s_vectors:
        for student_b,text_vector_b in s_vectors:
            if(student_a==student_b):
                pass
            else:
                progress = getProgressAsStringProgressBar(similarity(text_vector_a,text_vector_b)[0][1])
                #plagiarism_results.add((student_a,student_b,similarity(text_vector_a,text_vector_b)[0][1]))
                plagiarism_results.add((student_a,student_b,progress))
    r,c=1,1
    for i in plagiarism_results:
        c=1
        for j in range(3):
            w = tk.Label(window, text=i[j],fg='purple2',width=18,pady=8,padx=40,anchor='w',justify=LEFT, wraplength=300,font=('times', 20, 'bold'))
            w.grid(row=3+r,column=c)
            w.place(x=655,y=600,anchor='c')
            c+=1
        r+=1  
        
myButton = Button(window,text='Check Plagiarism',command=check_plagiarism ,fg='blue',bg='azure3', pady=20,padx=60,anchor='n',font=font1)
myButton.place(x=650,y=350,anchor='c')
                

window.mainloop()  # Keep the window open
```
