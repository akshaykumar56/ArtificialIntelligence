# Q1 What is the average price of each distinct brand listed?

```python
#there are 505 unique brand names in the dataset
brand.shape
```

```python
avg_price=[]
for i in brand:
    a=df_shoes['brand']==i
    rows=df_shoes[a]
    avg_price.append(rows['prices.amountMax'].mean())
```

```python
avg_price
```

```python
res = dict(zip(brand, avg_price))
```

```python
res
```

```python
res1=pd.DataFrame({'brand':brand,'avg_price':avg_price})
```

```python
res1
```

```python
res1.sort_values('avg_price',ascending=False,inplace=True)
```

```python
res1
```

```python
res1.head(10)
```

# Q2 Which brands have the highest prices?

```python
#for max_price for a certain brand
max_price=[]
for i in brand:
    a=df_shoes['brand']==i
    rows=df_shoes[a]
    max_price.append(rows['prices.amountMax'].max())
```

```python
max_price.sort()
```

```python
len(max_price)
```

```python
#creating dataframe
max_price_df=pd.DataFrame({'brand':brand,'max_price':max_price})
```

```python
max_price_df
```

```python
max_price_df.sort_values('max_price',ascending=False,inplace=True)
```

```python
max_price_df.head(10)
```

```python
no_of_items=[]
for i in brand:
    no_of_items.append((df_shoes['brand']==i).sum())
```
